```python
!pip install JPype1
!wget https://repo1.maven.org/maven2/org/apache/poi/poi/5.2.3/poi-5.2.3.jar -P /content/
!wget https://repo1.maven.org/maven2/org/apache/poi/poi-ooxml/5.2.3/poi-ooxml-5.2.3.jar -P /content/
!wget https://repo1.maven.org/maven2/org/apache/poi/poi-ooxml-schemas/5.2.3/poi-ooxml-schemas-5.2.3.jar -P /content/
!wget https://repo1.maven.org/maven2/org/apache/commons/commons-collections4/4.4/commons-collections4-4.4.jar -P /content/
!wget https://repo1.maven.org/maven2/org/apache/xmlbeans/xmlbeans/5.0.3/xmlbeans-5.0.3.jar -P /content/
# Download Apache Commons IO
!wget https://repo1.maven.org/maven2/commons-io/commons-io/2.11.0/commons-io-2.11.0.jar -P /content/

import jpype
import jpype.imports
from jpype.types import *

# Construct the classpath string
classpath = ':'.join([
    '/content/poi-5.2.3.jar',
    '/content/poi-ooxml-5.2.3.jar',
    '/content/poi-ooxml-schemas-5.2.3.jar',
    '/content/commons-collections4-4.4.jar',
    '/content/xmlbeans-5.0.3.jar',
    '/content/commons-io-2.11.0.jar'
])

if not jpype.isJVMStarted():
    # Start the JVM with the constructed classpath
    jpype.startJVM(classpath=classpath)
    print("JVM started")
# Define el código en Java
java_code = """
import org.apache.poi.xssf.usermodel.XSSFWorkbook;
import org.apache.poi.ss.usermodel.Sheet;
import org.apache.poi.ss.usermodel.Row;
import org.apache.poi.ss.usermodel.Cell;
import java.io.FileOutputStream;
import java.io.IOException;


public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("¡Hola Mundo!");
        // Crear un libro de Excel
        XSSFWorkbook workbook = new XSSFWorkbook();
        Sheet sheet = workbook.createSheet("Sheet1");
        Row row = sheet.createRow(0);
        Cell cell = row.createCell(0);
        cell.setCellValue("Hello, Apache POI!");

        // Guardar el libro de Excel
        try (FileOutputStream fileOut = new FileOutputStream("/content/example.xlsx")) {
            workbook.write(fileOut);
        } catch (IOException e) {
            e.printStackTrace();
        } finally {
            try {
                workbook.close();
            } catch (IOException e) {
                e.printStackTrace();
            }
        }
        System.out.println("Excel file created");
    }
}
"""
with open('HelloWorld.java', 'w') as f:
    f.write(java_code)


# Compila el código Java, including the commons-io JAR in the classpath
!javac -cp {classpath} HelloWorld.java

# Añade el directorio actual al classpath
jpype.addClassPath('.')


# Carga la clase compilada y ejecuta el método main
HelloWorld = jpype.JClass('HelloWorld')
HelloWorld.main(None)
```
