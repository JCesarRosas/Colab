```python
!pip install --upgrade selenium
# Instala Selenium

!pip install Pillow  # Install Pillow if you haven't already

# Descarga el driver de Chrome (o Firefox) según corresponda
!apt-get update
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

# Configura Selenium para que use el driver de Chrome
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from PIL import Image
import io

chrome_options = Options()
chrome_options.add_argument('--headless')  # Ejecución en modo headless (sin ventana)
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Inicia el navegador
driver = webdriver.Chrome(options=chrome_options)

# Abre YouTube
driver.get('https://www.youtube.com/')

# Imprime el título de la página actual
print(driver.title)

# Encuentra el campo de búsqueda
search_box = driver.find_element(By.NAME, 'search_query')

# Escribe "furiosa madmax" en el campo de búsqueda
search_box.send_keys('furiosa madmax')

# Presiona Enter para realizar la búsqueda
search_box.send_keys(Keys.RETURN)

# Espera un momento para que se cargue la página de resultados
driver.implicitly_wait(5)

# Imprime el título de la página actual
print(driver.title)

# Encuentra el primer resultado de búsqueda y haz clic en él
first_result = driver.find_element(By.CSS_SELECTOR, '#contents ytd-video-renderer h3 a')
first_result.click()

# Espera un momento para que se cargue el video
driver.implicitly_wait(5)

# Imprime el título de la página actual
print('Título del video:', driver.title)

# Imprime la URL del video
print('URL del video:', driver.current_url)

# Capture screenshot and convert it to bytes
screenshot_bytes = driver.get_screenshot_as_png()
image = Image.open(io.BytesIO(screenshot_bytes))

# Display the image
display(image)

# Cierra el navegador
driver.quit()
```
