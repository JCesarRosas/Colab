```python
!pip install JPype1==1.4.1
!wget https://repo1.maven.org/maven2/org/apache/commons/commons-math3/3.6.1/commons-math3-3.6.1.jar -P /content/

import jpype
import jpype.imports
from jpype.types import *

# Start the JVM (ensure this is called only once in the session)
if not jpype.isJVMStarted():
    jpype.startJVM(classpath=['/content/commons-math3-3.6.1.jar'])  # Start JVM if not already started
    print("Started JVM") # Print confirmation
else:
    print("JVM already running") # Print confirmation


# Import Java classes
from org.apache.commons.math3.stat.descriptive import DescriptiveStatistics

# Create a Java class instance
stats = DescriptiveStatistics()

# Use Java methods
stats.addValue(10.0)
stats.addValue(20.0)
stats.addValue(30.0)
result = stats.getMean()

print("Mean:", result)


# Define el código en Java
java_code = """
public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("¡Hola crayola!");
    }
}
"""

# Write Java code to a file
with open('HelloWorld.java', 'w') as f:
    f.write(java_code)

# Use ToolProvider to compile the Java code
import javax.tools
compiler = javax.tools.ToolProvider.getSystemJavaCompiler()
result = compiler.run(None, None, None, 'HelloWorld.java')

if result == 0:
    print("Compilation successful")
    # Add current directory to classpath
    jpype.addClassPath('.') # This line is added to fix the error
    HelloWorld = jpype.JClass('HelloWorld')
    HelloWorld.main(None) # Now you can load the class

else:
    print("Compilation failed")

# Now load the compiled class, adding the current directory '.' to the classpath
HelloWorld = jpype.JClass('HelloWorld')

# Ejecuta el método main
HelloWorld.main(None)


# Stop the JVM when done (ensure this is called only once at the end of the session)
# Commenting out the shutdown for now to allow other cells to use the JVM
# if jpype.isJVMStarted():
#     jpype.shutdownJVM()
```
