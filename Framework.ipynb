```python
#!pip install selenium
#!apt-get update
#!apt install chromium-chromedriver
#!cp /usr/lib/chromium-browser/chromedriver /usr/bin
#!pip uninstall -y pydrive
#!pip install pydrive
import pytest
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By


# Configuración de Selenium
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Ejecución en modo headless (sin interfaz gráfica)
options.add_argument('--no-sandbox') # This argument might help resolve the issue

@pytest.fixture(scope="module")
def driver():
    # Inicializar el driver
    driver = webdriver.Chrome(options=options) # Remove the 'chromedriver' argument if it's automatically detected
    yield driver
    # Cerrar el navegador después de todas las pruebas
    driver.quit()

def test_successful_login(driver):
    # Abrir la página de inicio de sesión
    driver.get('https://demo.guru99.com/test/newtours/index.php')
    # Encontrar el campo de usuario, contraseña y botón de inicio de sesión
    username_field = driver.find_element(By.NAME, 'userName')
    password_field = driver.find_element(By.NAME,'password')
    login_button = driver.find_element(By.NAME,'submit')
    # Ingresar credenciales
    username_field.send_keys('admin')
    password_field.send_keys('admin')
    # Hacer clic en el botón de inicio de sesión
    login_button.click()
    # Verificar si se ha iniciado sesión correctamente
    assert "Login" in driver.title

    # Obtener el título y la URL después de iniciar sesión
    page_title = driver.title
    page_url = driver.current_url

    return page_title, page_url

if __name__ == "__main__":
    # Crear un objeto de opciones de Chrome
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # Ejecución en modo headless (sin interfaz gráfica)
    options.add_argument('--no-sandbox') # This argument might help resolve the issue
    # Inicializar el driver con las opciones configuradas
    driver = webdriver.Chrome(options=options) # Remove the 'chromedriver' argument if it's automatically detected
    # Llamar a la función de prueba y pasar el driver como argumento
    page_title, page_url = test_successful_login(driver)
    # Imprimir los resultados
    print("Pruebas completadas.")
    print("Título de la página después del inicio de sesión:", page_title)
    print("URL de la página después del inicio de sesión:", page_url)
    # Cerrar el driver
    driver.quit()
```
