```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

# Configuración del driver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Para ejecutar en segundo plano sin abrir una ventana de navegador
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Inicializar el driver con las opciones configuradas
driver = webdriver.Chrome(options=chrome_options)

# Abrir la página de Google
driver.get("https://www.google.com")
print("Título de la página actual : " + driver.title)
time.sleep(3)

# Obtener el enlace de Gmail
gmail_icon = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div/div/div[1]/div/div/div[1]/div/div[1]/a")
#gmail_icon = driver.find_element(By.LINK_TEXT, "Gmail")

# Abrir el enlace de Gmail en una nueva ventana utilizando JavaScript
driver.execute_script("window.open(arguments[0], '_blank');", gmail_icon.get_attribute("href"))

# Esperar un momento para que la nueva ventana se abra completamente
time.sleep(3)

# Imprimir el número de manejadores de ventana disponibles
print("Número de manejadores de ventana disponibles:", len(driver.window_handles))

# Cambiar al manejador de la nueva ventana (Gmail)
driver.switch_to.window(driver.window_handles[1])

# Imprimir el título de la página actual (Gmail)
print("Título de la página actual : " + driver.title)

# Esperar un momento para la demostración
time.sleep(3)

# Imprimir el número de manejadores de ventana disponibles
print("Número de manejadores de ventana disponibles:", len(driver.window_handles))

# Cambiar de nuevo a la página de Google
driver.switch_to.window(driver.window_handles[0])

# Imprimir el título de la página actual (Google)
print("Título de la página actual : " + driver.title)

driver.switch_to.window(driver.window_handles[1])

# Imprimir el título de la página actual (Google)
print("Título de la página actual : " + driver.title)

driver.switch_to.window(driver.window_handles[0])

# Imprimir el título de la página actual (Google)
print("Título de la página actual : " + driver.title)

# Cerrar el navegador
driver.quit()
```
